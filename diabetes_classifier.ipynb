```python
# import numpy, matplotlib, etc.
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# sklearn imports
import sklearn
from sklearn import metrics
from sklearn import datasets
from sklearn import pipeline
from sklearn import linear_model
from sklearn import preprocessing
from sklearn import model_selection

# define plt settings
sns.set_theme()
plt.rcParams["font.size"] = 20
plt.rcParams["axes.labelsize"] = 20
plt.rcParams["xtick.labelsize"] = 20
plt.rcParams["ytick.labelsize"] = 20
plt.rcParams["legend.fontsize"] = 20
plt.rcParams["legend.markerscale"] = 1.5
plt.rcParams["figure.figsize"] = (20, 10)
plt.rcParams["legend.title_fontsize"] = 20
```

```python
diabetes = pd.read_csv('diabetes_data_upload.csv')
print(diabetes.columns)
```

```python
diabetes.head()
diabetes=diabetes.rename(columns={'class':'ill'})
diabetes=diabetes.replace(['No', 'Female', 'Negative', 'Yes', 'Male', 'Positive'],[0,0,0,1,1,1])
```

```python
feature_cols = ['Age','Gender','Polyuria','Polydipsia','sudden weight loss','weakness','Polyphagia','Genital thrush','visual blurring','Itching','Irritability','delayed healing','partial paresis','muscle stiffness','Alopecia','Obesity']
#feature_cols = ['Inches']
x = diabetes[feature_cols] # predictor
y = diabetes.ill # response
#Using the train_test_split to create train and test sets.
X_train, X_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.20)
```

```python
#Importing the Decision tree classifier from the sklearn library.
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
#Training the decision tree classifier. 
clf.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
